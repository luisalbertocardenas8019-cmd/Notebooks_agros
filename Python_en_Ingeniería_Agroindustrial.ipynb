## 3 Ejemplos sencillos de Python para Agroindustria

Aquí tienes tres fragmentos de código básicos que muestran cómo Python facilita la vida en esta carrera:

### 1. Cálculo del Índice de Madurez

Imagina que quieres automatizar el cálculo rápido de la relación entre sólidos solubles ($^\circ Brix$) y acidez para determinar si una fruta está lista.

```
# Datos de entrada
brix = 12.5
acidez_titulable = 0.8

# Cálculo del índice
indice_madurez = brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Fruta lista para cosecha.")
else:
    print("Estado: Falta maduración.")
```

### 2. Conversión de Unidades de Temperatura

En las plantas de procesamiento, es común recibir maquinaria o manuales en diferentes sistemas. Este script convierte grados Fahrenheit a Celsius para el control de un autoclave.

```
def fahrenheit_a_celsius(f):
    c = (f - 32) * 5/9
    return c

temp_f = 250
temp_c = fahrenheit_a_celsius(temp_f)

print(f"La temperatura del proceso es {temp_c:.2f}°C")
```

### 3. Estimación de Rendimiento de Extracción

Este código calcula cuánto aceite o jugo estás obteniendo realmente de una materia prima.

```
peso_materia_prima = 500  # kg (ej. aceitunas)
peso_aceite_obtenido = 95 # kg

rendimiento = (peso_aceite_obtenido / peso_materia_prima) * 100

print(f"Rendimiento del proceso: {rendimiento}%")

if rendimiento < 18:
    print("Alerta: Rendimiento por debajo del estándar.")
```
